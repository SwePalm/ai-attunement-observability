```python
import attunement.scenarios_handler as sh
import attunement.prompts_handler as ph
import attunement.litellm_handler as lh
import attunement.framework_handler as fh
```

## Define backend and model to use

```python
year_month = "2511"
backend = "nebius"
model_name = "meta-llama/Llama-3.3-70B-Instruct-fast"
```

## Get the scenarios that will be used as input

```python
model_name_in_file_name = model_name.split("/")[-1]
scenario_store = f"data/scenarios_{year_month}_{model_name_in_file_name}.json"

data_to_framework = sh.get_all_themes_data_to_framework(scenario_store)
print(data_to_framework.keys())
```

## Create a draft of the framework, defining the 5 waves

```python
create_framework_prompt = ph.get_framework_prompt(data_to_framework)
#print(create_framework_prompt)
```

```python
draft_framework = lh.get_text_from_litellm(create_framework_prompt, model_name= f"{backend}/{model_name}") 
print(draft_framework)
```

```python
# name this file same as scenarios, but this one framework year month and model used 
framework_store = f"data/framework_{year_month}_{model_name_in_file_name}.json"

fh.save_response_to_framework("draft", draft_framework, filepath=framework_store)
```

## Optional - Create Draft Report

```python
report_prompt = ph.get_report_prompt(draft_framework)
```

```python
report = lh.get_text_from_litellm(report_prompt, model_name= f"{backend}/{model_name}")
print(report)
```

```python
fh.save_response_to_framework("report", report, filepath=framework_store)
```
