### For each wave, find signals and classify them...to be used for velidate timelines
This was more struggle then expected, due to change for next iteration

```python
make_waves_list_prompt = ph.get_waves_list_prompt(draft_framework)
waves = gh.get_text_from_gemini(make_waves_list_prompt, model_name="gemini-2.5-flash")
print(waves)
```

How hard can it be to get a list...very it showed, you might need some skills here...

```python
inner_content = waves[2:-2]
waves = inner_content.split('", "')
```

```python
waves[10:-4].replace("\n","").replace("    ","")
```

```python
# Did struggle a bit to get good formatted python lists
import ast

try:
    waves = ast.literal_eval(waves[10:-4].replace("\n","").replace("    ",""))
except:
    print("Ops, no list")
len(waves)
```

Do NOT run the cells below if you don't have a list with 5 waves

```python
# If we got this far, we are good to go, but 

all_signals = {}

for i in range(0,len(waves)):
    wave = waves[i]
    print(f"{wave=}")
    this_prompt = ph.get_create_indicators_prompt(wave)
    this_response = gh.get_text_from_gemini(this_prompt, model_name="gemini-2.0-flash")
    wave_number=i+1
    all_signals[f"wave{wave_number}"] = this_response
print(len(all_signals))
```

```python
fh.save_response_to_framework("signals", all_signals, filepath=framework_store)
```
