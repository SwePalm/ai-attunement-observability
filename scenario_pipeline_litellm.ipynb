```python
import attunement.scenarios_handler as sh
import attunement.prompts_handler as ph
import attunement.litellm_handler as lh
import attunement.framework_handler as fh
```

```python
from attunement.config import THEMES_LIST
themes_list = THEMES_LIST
themes_list
```

```python
year_month = "2511"
backend = "nebius"
model_name = "meta-llama/Llama-3.3-70B-Instruct-fast"
```

```python
# name this file as scenarios year month and model used 
model_name_in_file_name = model_name.split("/")[-1]
scenario_store = f"data/scenarios_{year_month}_{model_name_in_file_name}.json"

sh.create_new_data_store(scenario_store)
```

```python
import time

for i in range(0, len(themes_list)):
    current_theme = themes_list[i]
    print(current_theme)
    # Step 0, Define Question
    step_start_time = time.time()
    current_explore_question_prompt = ph.get_explore_question_prompt(current_theme)
    current_theme_question = lh.get_text_from_litellm(current_explore_question_prompt, model_name= f"{backend}/{model_name}")
    sh.save_response_to_theme(current_theme, "question", current_theme_question, filepath=scenario_store)
    print(f"Step 0 Duration: {time.time() - step_start_time:.4f} seconds\n")
    
    # Step 1, Exploration
    step_start_time = time.time()
    current_theme_exploration_prompt = ph.get_theme_exploration_prompt(current_theme, current_theme_question)
    theme_exploration_text = lh.get_text_from_litellm(current_theme_exploration_prompt, model_name= f"{backend}/{model_name}")
    sh.save_response_to_theme(current_theme, "exploration", theme_exploration_text, filepath=scenario_store)
    print(f"Step 1 Duration: {time.time() - step_start_time:.4f} seconds\n")

    # Step 2, PESTLE
    step_start_time = time.time()
    pestle_analysis_prompt = ph.get_pestle_analysis_prompt(theme_exploration_text)
    pestle_analysis_text = lh.get_text_from_litellm(pestle_analysis_prompt, model_name= f"{backend}/{model_name}")
    sh.save_response_to_theme(current_theme, "pestle", pestle_analysis_text, filepath=scenario_store)
    print(f"Step 2 Duration: {time.time() - step_start_time:.4f} seconds\n")

    # Step 3, Forces and Feelings
    step_start_time = time.time()
    forces_feelings_prompt = ph.get_forces_feelings_prompt(theme_exploration_text, pestle_analysis_text)
    forces_feelings_text = lh.get_text_from_litellm(forces_feelings_prompt, model_name= f"{backend}/{model_name}")
    sh.save_response_to_theme(current_theme, "forces_feelings", forces_feelings_text, filepath=scenario_store)
    print(f"Step 3 Duration: {time.time() - step_start_time:.4f} seconds\n")

    # Step 4, Generate Scenario
    step_start_time = time.time()
    scenario_prompt = ph.get_scenario_prompt(theme_exploration_text, pestle_analysis_text, forces_feelings_text)
    scenario_text = lh.get_text_from_litellm(scenario_prompt, model_name= f"{backend}/{model_name}")
    sh.save_response_to_theme(current_theme, "scenario", scenario_text, filepath=scenario_store)
    print(f"Step 4 Duration: {time.time() - step_start_time:.4f} seconds\n")

```
